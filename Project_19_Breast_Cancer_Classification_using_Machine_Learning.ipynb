```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your cleaned data
df = pd.read_excel('cleaned_data.xlsx', engine='openpyxl')

# Drop non-essential columns
employment_df = df.drop(columns=['gender', 'imp_cat'])

# Ensure that all data in employment_df is numeric
employment_df = employment_df.apply(pd.to_numeric, errors='coerce')

# Sum the values for each employment category
employment_sums = employment_df.sum()

# Find the top 5 employment categories
top_employment_categories = employment_sums.nlargest(5).index

# Filter the DataFrame to include only the top 5 employment categories
df_top_employment = df[['gender', 'imp_cat'] + list(top_employment_categories)]

# Create a pivot table for the heatmap
plt.figure(figsize=(12, 8))

# Iterate over top 5 employment categories to create heatmaps
for i, category in enumerate(top_employment_categories):
    plt.subplot(3, 2, i + 1)  # Arrange the heatmaps in a 3x2 grid
    heatmap_data = df_top_employment.pivot_table(index='gender', columns='imp_cat', values=category, aggfunc='sum', fill_value=0)
    sns.heatmap(heatmap_data, annot=True, cmap="Reds", fmt="d", annot_kws={'size': 10})
    plt.title(f'Heatmap for {category}', fontsize=12)
    plt.xlabel('Impairment Categories', fontsize=10)
    plt.ylabel('Gender', fontsize=10)

plt.tight_layout()
plt.savefig('heatmap2.png')
plt.show()
```
